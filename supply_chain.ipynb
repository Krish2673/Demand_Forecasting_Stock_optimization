```python
def forecast_product(item_id, store_id, sales_df, calendar_df, holidays_df=None):
    from prophet import Prophet

    sample_id = f"{item_id}_{store_id}_validation"
    ts = sales_df[sales_df['id'] == sample_id].iloc[:, 6:].T
    print(f"[{sample_id}] Time Series shape: {ts.shape}")
    ts.columns = ['y']
    ts['ds'] = pd.date_range(start='2011-01-29', periods=len(ts))
    ts.reset_index(drop=True, inplace=True)

    recent_sales = ts['y'].tail(28).mean()
    stock_level = recent_sales * 7

    if holidays_df is not None:
        model = Prophet(holidays=holidays_df, yearly_seasonality=True, weekly_seasonality=True)
    else:
        model = Prophet(yearly_seasonality=True, weekly_seasonality=True)
    model.fit(ts)

    future = model.make_future_dataframe(periods=28)
    forecast = model.predict(future)
    forecast_df = forecast[['ds', 'yhat']].tail(28).copy()

    stock = stock_level
    alerts = []
    for pred in forecast_df['yhat']:
        if pred > stock:
            alerts.append("⚠️ Understock")
        elif pred < stock * 0.3:
            alerts.append("📦 Overstock")
        else:
            alerts.append("✅ OK")
        stock -= pred
    forecast_df['stock_alert'] = alerts

    calendar_df['date'] = pd.to_datetime(calendar_df['date'])
    merged = forecast_df.merge(calendar_df[['date', 'event_name_1']], left_on='ds', right_on='date', how='left')
    merged['seasonal_flag'] = merged['event_name_1'].apply(lambda x: "🎉 Seasonal Spike" if pd.notnull(x) else "")
    merged.drop(columns=['date', 'event_name_1'], inplace=True)

    merged['item_id'] = item_id
    merged['store_id'] = store_id

    return merged
```

```python
id_cols = ['id', 'item_id', 'cat_id', 'store_id']
sales_cols = [col for col in sales.columns if col.startswith('d_')]
subset = sales[id_cols + sales_cols]

subset['active_days'] = subset[sales_cols].gt(0).sum(axis=1)

top_5_foods = subset[subset['cat_id'] == 'FOODS'].nlargest(5, 'active_days')
top_5_hobbies = subset[subset['cat_id'] == 'HOBBIES'].nlargest(5, 'active_days')
top_5_household = subset[subset['cat_id'] == 'HOUSEHOLD'].nlargest(5, 'active_days')

final_selection = pd.concat([top_5_foods, top_5_hobbies, top_5_household], ignore_index=True)

final_selection[['id', 'item_id', 'cat_id', 'active_days', 'store_id']]
```

```python
products = list(zip(final_selection['item_id'], final_selection['store_id']))
```

```python
unique_products = {}
for item_id, store_id in products:
    if item_id not in unique_products:
        unique_products[item_id] = store_id

products = [(item, store) for item, store in unique_products.items()]
```

```python
products
```

```python
results = []

for item_id, store_id in products:
    print(f"Forecasting for {item_id} at {store_id}...")
    df = forecast_product(item_id, store_id, sales, calendar, holidays)
    results.append(df)

final_forecast_df = pd.concat(results, ignore_index=True)
```

```python
product_name_map = {
    "FOODS_3_586": "Whole Grain Bread",
    "FOODS_3_555": "Oatmeal Choco-Chip Cookies",
    "FOODS_3_252": "Dark Chocolate Bar",
    "FOODS_3_080": "Greek Yogurt",
    
    "HOBBIES_1_074": "Sketchbook & Drawing Pencil Kit",
    "HOBBIES_1_254": "Watercolor Painting Set",
    "HOBBIES_1_323": "Portable Chess & Board Game Set",
    "HOBBIES_1_337": "Advanced Lego Robotics Kit",
    
    "HOUSEHOLD_1_072": "Soft Cotton Bath Towel",
    "HOUSEHOLD_1_118": "Aromatic Room Freshener",
    "HOUSEHOLD_1_198": "Multipurpose Cleaning Spray",
    "HOUSEHOLD_1_373": "Eco-Friendly Dishwashing Liquid",
}
```

```python
final_forecast_df["product_name"] = final_forecast_df["item_id"].map(product_name_map)
```

```python
cols = ["product_name", "item_id", "store_id", "date", "predicted_sales", "stock_alert", "seasonal_flag"]
final_forecast_df = final_forecast_df[cols]
```

```python
import json

def convert_forecast_to_json(df):
    grouped_data = []

    for (item_id, store_id), group in df.groupby(["item_id", "store_id"]):
        product_name = product_name_map.get(item_id, item_id)

        product_forecast = {
            "product_name": product_name,
            "item_id": item_id,
            "store_id": store_id,
            "forecasts": []
        }

        for _, row in group.iterrows():
            product_forecast["forecasts"].append({
                "date": row["date"].strftime("%Y-%m-%d"),
                "predicted_sales": round(row["predicted_sales"], 2),
                "stock_alert": row["stock_alert"],
                "seasonal_flag": row["seasonal_flag"]
            })

        grouped_data.append(product_forecast)

    return grouped_data

grouped_json = convert_forecast_to_json(final_forecast_df)

with open("dashboard_data.json", "w") as f:
    json.dump(grouped_json, f, indent=2)

print("✅ JSON export complete: dashboard_data.json")
```
